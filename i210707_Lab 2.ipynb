```python
print("-----------------------------------------------------------------------------------------------------------------------")
print("\t\t\t\t\t\t Task 3")
print("-----------------------------------------------------------------------------------------------------------------------")


class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price
    
    def display_details(self):
        print("Product ID:", self.product_id)
        print("Name:", self.name)
        print("Price:", self.price)
        
    def apply_discount(self, discount_percentage):
        return (self.price * discount_percentage / 100)
        

class Electronics(Product):
    def __init__(self, product_id, name, price, brand):
        super().__init__(product_id, name, price)
        self.brand = brand
    
    def display_details(self):
        super().display_details()
        print("Brand:", self.brand)
        

class Clothing(Product):
    def __init__(self, product_id, name, price, size):
        super().__init__(product_id, name, price)
        self.size = size
    
    def display_details(self):
        super().display_details()
        print("Size:", self.size)
         

electronics_item = Electronics("E1001", "Laptop", 1200, "Dell")
electronics_item_2 = Electronics("E1002", "Smartphone", 800, "Samsung")
electronics_item_3 = Electronics("E1003", "Headphones", 50, "Sony")


clothing_item = Clothing("C2001", "T-Shirt", 25, "Medium")
clothing_item_2 = Clothing("C2002", "Jeans", 50, "32")
clothing_item_3 = Clothing("C2003", "Dress", 80, "Small")


cart = [electronics_item,electronics_item_2,electronics_item_3, clothing_item,clothing_item_2,clothing_item_3]

total_price = 0
for item in cart:
    print("Item Details:")
    item.display_details()
    total_price += item.price
    print("-------------------------")
    
print("Total Price:", total_price)
    
print ("DO YOU WANT A DISCOUNT : ")
    
text = input("Enter 'y' if you want a discount: ")
    
if text == 'y':
    a = clothing_item.apply_discount(40)  

total_price -=a
    
print("Total Price:", total_price)
```
