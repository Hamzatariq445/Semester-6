```python
#Question 2
def display_players(players):
    print("Player Information:")
    for player in players:
        print("Name:", player[0],"  Goals: ",player[1], "Assists: ",player[2],"Playing Time ",player[3])

def sort_players(players, metric_index):
    n = len(players)

    for i in range(n):
        for j in range(0, n - i - 1):
            if players[j][metric_index] < players[j + 1][metric_index]:
                players[j], players[j + 1] = players[j + 1], players[j]

players = [
    ("Babar", 10, 5, 90),
    ("Rizwan", 8, 6, 75),
    ("Abdullah", 15, 3, 85)
]

print("Original Player Information:")
display_players(players)

print("\nSorted Bassed On Goals")

sort_players(players, 1)
display_players(players)

print("\nSorted Bassed On Assits")

sort_players(players, 2)
display_players(players)


print("\nSorted Bassed On Playing Time")

sort_players(players, 3)
display_players(players)
```

```python
class Route:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

class Schedule:
    def __init__(self, departure_time):
        self.departure_time = departure_time

    def __str__(self):
        return f"Departure Time: {self.departure_time}"

class Vehicle:
    def __init__(self, name):
        self.name = name
        self.schedule_ids = []  # Store assigned schedule IDs
        self.route_ids = []     # Store assigned route IDs

    def add_schedule(self, schedule):
        self.schedule_ids.append(schedule.departure_time)  # Append only the schedule ID

    def add_route(self, route):
        self.route_ids.append(route)

    def display(self):
        print(f"Vehicle: {self.name}")
        if self.schedule_ids:
            print("Assigned Schedule IDs:", ", ".join(str(schedule_id) for schedule_id in self.schedule_ids))
        if self.route_ids:
            print("Assigned Route IDs:", ", ".join(str(route) for route in self.route_ids))
        print("\n")

class Transport:
    def __init__(self):
        self.routes = []
        self.schedules = []
        self.vehicles = []

    def add_route(self, route):
        self.routes.append(route)

    def add_schedule(self, schedule):
        self.schedules.append(schedule)

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def get_routes(self):
        return self.routes

    def get_schedules(self):
        return self.schedules

    def get_vehicles(self):
        return self.vehicles

route1 = Route("1")
route2 = Route("2")

schedule1 = Schedule(" 1 Timming 10:00 AM -- 12:00 PM")
schedule2 = Schedule(" 2 Timming 12:00 PM -- 2:00 PM")

vehicle1 = Vehicle("Vehicle 1")
vehicle2 = Vehicle("Vehicle 2")
vehicle1.add_schedule(schedule1)
vehicle2.add_schedule(schedule2)

vehicle1.add_route(route1)
vehicle1.add_route(route2)
vehicle2.add_route(route2)

transport = Transport()

transport.add_vehicle(vehicle1)
transport.add_vehicle(vehicle2)

transport.add_route(route1)
transport.add_route(route2)

transport.add_schedule(schedule1)
transport.add_schedule(schedule2)

print("Vehicles:\n")
for vehicle in transport.get_vehicles():
    vehicle.display()
```
